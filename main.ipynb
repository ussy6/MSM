```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize # カラーバーの描画に用いる
import cartopy.crs as ccrs
import datetime
```

```python
for day in [27]: # range(22,28+1)
    # 描画したい年・月
    draw_year = 2022
    draw_month = 12
    draw_day = day
    draw_hour = 0

    # vminはカラーバーの下限，vmaxはカラーバーの上限
    # vintはカラーバーの間隔
    vmin = -20
    vmax = 20
    vint = 0.1

    # 深い青から深い赤に向かうカラーバーを指定
    cm = plt.get_cmap('seismic') 

    loadfile = f'npz/msm_2022{draw_month}{draw_day}_s.npz' # 入力ファイル名を定義
    sst_dataset = np.load(loadfile) # データセットはまずデータセットごと入力します
    dataname_list = sst_dataset.files

    sst = sst_dataset[dataname_list[1]] # 海面水温(sea surface temperature)を変数sstに保存
    pres = sst_dataset[dataname_list[4]] # 海面水温(sea surface temperature)を変数sstに保存
    lon2 = sst_dataset['lon2'] # 経度(longitude)を変数lon2に保存（2は「2次元配列」の意味）
    lat2 = sst_dataset['lat2'] # 緯度(latitude)を変数lat2に保存
    y = sst_dataset['y'] # 年(year)を変数yに保存
    m = sst_dataset['m'] # 月(month)を変数mに保存
    d = sst_dataset['d']
    h = sst_dataset['h']

    # for i in range(0,24,3): # range(h.size)
    for i in [21]: # range(h.size)
        draw_hour = i

        # 色で塗られた等高線を描く
        cs = plt.contourf(lon2, lat2, \
                        np.squeeze(sst[:, :, (y==draw_year)*(m==draw_month)*(d==draw_day)*(h==draw_hour)]-273.15), \
                        cmap=cm, norm=Normalize(vmin=vmin, vmax=vmax),\
                        levels=np.arange(vmin,vmax+vint,vint), extend='both')  
        p1 = plt.contour(lon2, lat2,
                        np.squeeze(pres[:, :, (y==draw_year)*(m==draw_month)*(d==draw_day)*(h==draw_hour)]*0.01),
                        levels=np.arange(960, 1040, 4), colors=["gray"], linewidths=[0.5,0.5,0.5,1])
                        # levels=np.arange(950,1050,2))
        p1.clabel(fmt="%d", fontsize=7)
        # \はプログラムが長い時に用いる改行なので，一行で書くなら必要なし
        # np.squeezeは，360x180x1の配列を360x180の配列として認識させるための関数

        plt.colorbar(cs, ticks=list(range(-20,20,2))) # カラーバーをつける
        plt.xlabel('Longitude') # 横軸のラベル
        plt.ylabel('Latitude') # 縦軸のラベル
        title = datetime.datetime(int(y[i]),int(m[i]),int(d[i]),int(draw_hour))
        # タイトルの文字列をanom_titleに代入。strは，数字を文字列に直している
        plt.title(title) # タイトルをつける
        plt.xlim(120, 150) # 横軸の範囲
        plt.ylim(22.39999962, 47.59999847) # 縦軸の範囲
        plt.savefig(f"2022{draw_month}{day}_{i}.jpg")
        plt.savefig(f"2022{draw_month}{day}_{i}.svg")
        plt.show() # ここまでの条件で描画（セル末尾の実行のみ省略可能）
```
